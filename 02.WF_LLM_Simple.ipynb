```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
import os
```

```python
load_dotenv()
```

```python
model = ChatOpenAI()
```

```python
# Create state type
class LLMState(TypedDict):
    question: str
    answer: str
```

```python
def llm_qa(state: LLMState) -> LLMState:

    # Extract the question from state 
    question = state['question']
    # From a prompt
    prompt =  f"Answer the following question {question}"
    # ask that question to the LLm
    answer = model.invoke(prompt).content
    # Update the answer in the state
    state['answer'] = answer
    
    return state
```

```python
# Build the graph
graph = StateGraph(LLMState)

# Add Nodes
graph.add_node('llm_qa', llm_qa)
# Add Edges
graph.add_edge(START, 'llm_qa')
graph.add_edge('llm_qa', END)

workflow = graph.compile()
```

```python
# Execute

initial_state = {'question': 'How far is moon from the earth?'}
final_state = workflow.invoke(initial_state)
print(final_state)
```

```python
# Compile and execute
workflow = graph.compile()
initial_state = {'question': 'How far is the moon from the Earth?'}
final_state = workflow.invoke(initial_state)
# Compile
print(final_state['answer']) # final_state['answer'] to see only answer
```

```python
# We can get the answer by invoking llm only
model.invoke('How far is the moon from the Earth?').content
# but our end goal is different
```
